```python
# https://habr.com/ru/articles/345502/
from subprocess import Popen, PIPE, check_output
import re
import json
import sys
import hashlib

path = '\"C:\\Program Files\\smartmontools\\bin\\smartctl\"'  # for linux use 'sudo smartctl'
codec = 'windows-1252'  # for linux use utf8

data = check_output(path + " --scan").decode(codec)
disks = set(re.findall(r'/dev/(.{3})', data))
output = []

for disk in disks:
    all_smart = Popen('{} -x /dev/{}'.format(path, disk), shell=True, stdout=PIPE).communicate()[0].decode(codec)
    output.append({"DISK_NAME": disk, "PARAMS": all_smart})
output = {"data": output}
print(json.dumps(output))
```
